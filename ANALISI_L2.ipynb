```python
df_corsi = pd.read_csv(
    os.path.join(BASE, "MURiscrittixcorsodistudi.csv"),
    sep=None, engine="python", encoding="latin-1"
)

# Filtra L-2 e anni 2021-2024
anni = ["2021/2022", "2022/2023", "2023/2024", "2024/2025"]

df_l2_corsi = df_corsi[
    (df_corsi["ClasseNUMERO"] == "L-2") &
    (df_corsi["AnnoA"].isin(anni))
].copy()

print(f"Righe trovate: {len(df_l2_corsi)}")
print()
print("Atenei e corsi trovati:")
print(
    df_l2_corsi.groupby(["AteneoNOME", "CorsoNOME"])["Isc"]
    .sum()
    .sort_values(ascending=False)
    .to_string()
)
```

```python
# Totale iscritti L-2 per ateneo e anno
df_l2_anno = (
    df_l2_corsi.groupby(["AnnoA", "AteneoNOME"])["Isc"]
    .sum()
    .reset_index()
)

# Quanti atenei attivi per anno
print("Atenei attivi per anno:")
print(df_l2_corsi.groupby("AnnoA")["AteneoNOME"].nunique().to_string())
print()
print("Iscritti totali per anno:")
print(df_l2_corsi.groupby("AnnoA")["Isc"].sum().to_string())
```

```python
# Dizionario ateneo → regione
ATENEO_REGIONE = {
    "Torino":                  "Piemonte",
    "Piemonte Orientale":      "Piemonte",
    "Genova":                  "Liguria",
    "Milano":                  "Lombardia",
    "Milano Bicocca":          "Lombardia",
    "Milano San Raffaele":     "Lombardia",
    "Insubria":                "Lombardia",
    "Brescia":                 "Lombardia",
    "Pavia":                   "Lombardia",
    "Trento":                  "Trentino",
    "Verona":                  "Veneto",
    "Padova":                  "Veneto",
    "Trieste":                 "Friuli",
    "Udine":                   "Friuli",
    "Bologna":                 "Emilia",
    "Ferrara":                 "Emilia",
    "Modena e Reggio Emilia":  "Emilia",
    "Parma":                   "Emilia",
    "Firenze":                 "Toscana",
    "Siena":                   "Toscana",
    "Pisa":                    "Toscana",
    "Perugia":                 "Umbria",
    "Camerino":                "Marche",
    "Urbino":                  "Marche",
    "Roma La Sapienza":        "Lazio",
    "Roma Tor Vergata":        "Lazio",
    "Tuscia":                  "Lazio",
    "L'Aquila":                "Abruzzo",
    "Teramo":                  "Abruzzo",
    "Napoli Federico II":      "Campania",
    "Napoli Vanvitelli":       "Campania",
    "Sannio":                  "Campania",
    "Bari":                    "Puglia",
    "Salento":                 "Puglia",
    "Foggia":                  "Puglia",
    "Basilicata":              "Basilicata",
    "Catanzaro":               "Calabria",
    "Calabria":                "Calabria",
    "Palermo":                 "Sicilia",
    "Catania":                 "Sicilia",
    "Messina":                 "Sicilia",
    "Cagliari":                "Sardegna",
    "Sassari":                 "Sardegna",
}

REGIONE_MACRO = {
    "Piemonte":   "Nord",
    "Liguria":    "Nord",
    "Lombardia":  "Nord",
    "Trentino":   "Nord",
    "Veneto":     "Nord",
    "Friuli":     "Nord",
    "Emilia":     "Nord",
    "Toscana":    "Centro",
    "Umbria":     "Centro",
    "Marche":     "Centro",
    "Lazio":      "Centro",
    "Abruzzo":    "Sud",
    "Campania":   "Sud",
    "Puglia":     "Sud",
    "Basilicata": "Sud",
    "Calabria":   "Sud",
    "Sicilia":    "Isole",
    "Sardegna":   "Isole",
}

df_l2_corsi["regione"]    = df_l2_corsi["AteneoNOME"].map(ATENEO_REGIONE)
df_l2_corsi["macro_area"] = df_l2_corsi["regione"].map(REGIONE_MACRO)

# Verifica
non_mappati = df_l2_corsi[df_l2_corsi["regione"].isna()]["AteneoNOME"].unique()
if len(non_mappati) > 0:
    print(f"⚠️  Atenei non mappati: {non_mappati}")
else:
    print("✅ Tutti gli atenei mappati correttamente!")

print()
print("Iscritti per macro-area (media 2021-2024):")
print(
    df_l2_corsi.groupby("macro_area")["Isc"]
    .sum()
    .sort_values(ascending=False)
    .to_string()
)
```

```python
# Iscritti per macro-area per anno
df_macro_anno = (
    df_l2_corsi.groupby(["AnnoA", "macro_area"])["Isc"]
    .sum()
    .reset_index()
)

print("Iscritti per macro-area per anno:")
print(df_macro_anno.pivot(index="AnnoA", columns="macro_area", values="Isc").to_string())
print()

# Totale per anno come controllo
print("Totale per anno:")
print(df_macro_anno.groupby("AnnoA")["Isc"].sum().to_string())
```

```python
# Totale iscritti per regione per anno (somma tutti gli anni)
df_regione = (
    df_l2_corsi.groupby(["AnnoA", "regione", "macro_area", "AteneoNOME", "CorsoNOME"])["Isc"]
    .sum()
    .reset_index()
)

# Per la mappa: totale iscritti per regione (somma tutti gli anni 2021-2024)
df_mappa = (
    df_l2_corsi.groupby(["regione", "macro_area"])["Isc"]
    .sum()
    .reset_index()
    .rename(columns={"Isc": "iscritti_totale"})
)

# Per l'hover: lista corsi per regione
df_hover = (
    df_l2_corsi.groupby(["regione", "AteneoNOME", "CorsoNOME"])["Isc"]
    .sum()
    .reset_index()
)

# Costruisce testo hover per regione
def crea_hover(regione):
    subset = df_hover[df_hover["regione"] == regione].sort_values("Isc", ascending=False)
    testo = f"<b>{regione}</b><br><br>"
    for _, row in subset.iterrows():
        testo += f"🎓 {row['AteneoNOME']}<br>"
        testo += f"   {row['CorsoNOME']}: {row['Isc']:,} iscritti<br>"
    return testo

df_mappa["hover_text"] = df_mappa["regione"].apply(crea_hover)

print("✅ Dati pronti per la mappa!")
print(df_mappa[["regione", "macro_area", "iscritti_totale"]].to_string(index=False))
```

```python
# Dati per mappa con filtro anno
df_mappa_anni = (
    df_l2_corsi.groupby(["AnnoA", "regione", "macro_area"])["Isc"]
    .sum()
    .reset_index()
    .rename(columns={"Isc": "iscritti_totale"})
)

# Hover dettagliato per regione e anno
df_hover_anni = (
    df_l2_corsi.groupby(["AnnoA", "regione", "AteneoNOME", "CorsoNOME"])["Isc"]
    .sum()
    .reset_index()
)

def crea_hover_anno(regione, anno):
    subset = df_hover_anni[
        (df_hover_anni["regione"] == regione) &
        (df_hover_anni["AnnoA"] == anno)
    ].sort_values("Isc", ascending=False)
    testo = f"<b>{regione}</b><br>Anno: {anno}<br><br>"
    for _, row in subset.iterrows():
        testo += f"🎓 {row['AteneoNOME']}<br>"
        testo += f"&nbsp;&nbsp;{row['CorsoNOME']}: {int(row['Isc']):,} iscritti<br>"
    return testo

df_mappa_anni["hover_text"] = df_mappa_anni.apply(
    lambda r: crea_hover_anno(r["regione"], r["AnnoA"]), axis=1
)

# Verifica
print("✅ Dati pronti!")
print()
print("Anni disponibili:", df_mappa_anni["AnnoA"].unique().tolist())
print()
print("Esempio hover Lazio 2024/2025:")
print(crea_hover_anno("Lazio", "2024/2025"))
```

```python
# ── Dizionario geo ────────────────────────────────────────────────────────────
REGIONE_GEO = {
    "Piemonte":   "Piemonte",
    "Liguria":    "Liguria",
    "Lombardia":  "Lombardia",
    "Trentino":   "Trentino-Alto Adige/Südtirol",
    "Veneto":     "Veneto",
    "Friuli":     "Friuli-Venezia Giulia",
    "Emilia":     "Emilia-Romagna",
    "Toscana":    "Toscana",
    "Umbria":     "Umbria",
    "Marche":     "Marche",
    "Lazio":      "Lazio",
    "Abruzzo":    "Abruzzo",
    "Campania":   "Campania",
    "Puglia":     "Puglia",
    "Basilicata": "Basilicata",
    "Calabria":   "Calabria",
    "Sicilia":    "Sicilia",
    "Sardegna":   "Sardegna",
}

df_mappa_anni["reg_name"] = df_mappa_anni["regione"].map(REGIONE_GEO)

GEOJSON_URL = (
    "https://raw.githubusercontent.com/openpolis/geojson-italy/"
    "master/geojson/limits_IT_regions.geojson"
)

# ── Hover aggiornato con totale ───────────────────────────────────────────────
def crea_hover_anno(regione, anno):
    subset = df_hover_anni[
        (df_hover_anni["regione"] == regione) &
        (df_hover_anni["AnnoA"] == anno)
    ].sort_values("Isc", ascending=False)

    totale = int(subset["Isc"].sum())

    testo = f"<b>{regione}</b><br>"
    testo += f"Anno: {anno} | Totale iscritti: <b>{totale:,}</b><br><br>"
    for _, row in subset.iterrows():
        testo += f"🎓 {row['AteneoNOME']}<br>"
        testo += f"&nbsp;&nbsp;{row['CorsoNOME']}: {int(row['Isc']):,} iscritti<br>"
    return testo

df_mappa_anni["hover_text"] = df_mappa_anni.apply(
    lambda r: crea_hover_anno(r["regione"], r["AnnoA"]), axis=1
)

# ── Mappa ─────────────────────────────────────────────────────────────────────
anni_ordinati = ["2021/2022", "2022/2023", "2023/2024", "2024/2025"]
no_corso_nomi = ["Valle d'Aosta/Vallée d'Aoste", "Molise"]

z_min = df_mappa_anni["iscritti_totale"].min()
z_max = df_mappa_anni["iscritti_totale"].max()

fig = go.Figure()

for i, anno in enumerate(anni_ordinati):
    subset = df_mappa_anni[df_mappa_anni["AnnoA"] == anno]

    fig.add_trace(go.Choropleth(
        geojson=GEOJSON_URL,
        locations=subset["reg_name"],
        featureidkey="properties.reg_name",
        z=subset["iscritti_totale"],
        colorscale="Blues",
        zmin=z_min,
        zmax=z_max,
        colorbar=dict(title="Iscritti", x=1.0),
        marker_line_color="white",
        marker_line_width=1,
        text=subset["hover_text"],
        hovertemplate="%{text}<extra></extra>",
        name=anno,
        visible=(i == 0),
    ))

    fig.add_trace(go.Choropleth(
        geojson=GEOJSON_URL,
        locations=no_corso_nomi,
        featureidkey="properties.reg_name",
        z=[0, 0],
        colorscale=[[0, "#4B5563"], [1, "#4B5563"]],
        showscale=False,
        marker_line_color="white",
        marker_line_width=1,
        hovertemplate="<b>%{location}</b><br>Nessun corso L2 attivo<extra></extra>",
        name="Nessun corso",
        visible=(i == 0),
    ))

# ── Bottoni filtro anno ───────────────────────────────────────────────────────
buttons = []
for i, anno in enumerate(anni_ordinati):
    visibilita = []
    for j in range(len(anni_ordinati)):
        visibilita.append(j == i)
        visibilita.append(j == i)

    buttons.append(dict(
        label=anno,
        method="update",
        args=[
            {"visible": visibilita},
            {"title": f"🗺️ Iscritti L-2 Biotecnologie per Regione — {anno}"}
        ]
    ))

fig.update_layout(
    title=f"🗺️ Iscritti L-2 Biotecnologie per Regione — {anni_ordinati[0]}",
    updatemenus=[dict(
        type="buttons",
        direction="right",
        x=0.5,
        xanchor="center",
        y=1.08,
        yanchor="top",
        buttons=buttons,
        bgcolor="white",
        bordercolor="#2563EB",
        borderwidth=1,
        font=dict(size=12),
    )],
    margin={"r": 0, "t": 80, "l": 0, "b": 0},
    height=600,
    font_family="Georgia",
    title_font_size=18,
)

fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
